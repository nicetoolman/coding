```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()

# 创建数据集
os.makedirs('data', exist_ok=True)#创建一个名为data的文件夹，如果已经存在，则不创建

# 下载数据集
dataset_name = "uciml/red-wine-quality-cortez-et-al-2009"
api.dataset_download_files(dataset_name, path='data', unzip=True)#下载数据集，并解压
```

```python
import pandas as pd

red_wine = pd.read_csv('data/winequality-red.csv')
red_wine.head()
```

```python
red_wine.shape #查看数据集的(行数, 列数)
```

```python
red_wine.describe() #查看数据集的统计信息，包括均值、标准差、最小值、最大值、四分位数等
```

```python
red_wine.info() #查看数据集的类型信息,包括数据类型、非空值数量、内存使用情况等
```

```python
red_wine.isna().sum() #查看数据集的缺失值数量
#语法解析：.isna()用于判断数据集中的缺失值，返回一个布尔值的DataFrame，True表示缺失值，False表示非缺失值。.sum()用于计算布尔值的DataFrame中True的数量，即缺失值的数量。 
```

```python
input_shape = [11]
output_shape = [1]

from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(1, input_shape=input_shape)
])

model.summary() #查看模型结构,包括层数、每层的参数数量、总参数数量等
```

```python
model.weights #查看模型权重

```

```python
# 转换为 numpy 数组查看具体值
for w in model.weights:
    print(f"名称: {w.name}")
    print(f"形状: {w.shape}")
    print("数值:")
    print(w.numpy())
    print("\n")
```
